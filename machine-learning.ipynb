```python
#feature columns having numerical data only and categorical column to be changed to numerical equivalent
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population', 'Price']]
```

```python
#target column
y = df['Price']
```

```python
#split the model for training for automation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
#coefficient data frame - used to check the impact of per unit changes
cdf = pd.DataFrame(lm.coef_, X_train.columns, columns=['Coeff'])
```

```python
#predictions on X data set
predictions = lm.predict(X_test)
```

```python
#check the deviation of y_test from predictions as histogram model
sns.distplot(y_test - predictions, bins=50)
```

```python
#loss function to reduce the errors into predictions
from sklearn import metrics
metrics.mean_absolute_error(y_test, predictions)
```

```python
metrics.mean_squared_error(y_test, predictions)
```

```python
#root mean squared error - RMSE
np.sqrt(metrics.mean_squared_error(y_test, predictions))
```

```python
#training and testing data
customers.columns
```

```python
X = customers[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership', 'Yearly Amount Spent']]
```

```python
y = customers['Yearly Amount Spent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# train fit the model
lm.fit(X_train, y_train)
```

```python
#get the coefficients
lm.coef_
```

```python
#predicts based on X_test
predictions = lm.predict(X_test)
```

```python
#create a scatterd plot of the real test values Vs predicted values
plt.scatter(y_test, predictions)
plt.xlabel('Y Test (True Value)')
plt.ylabel('Predicted Values')
```

```python
#The modeling is doing quite well, as it is linear graph with very low residuals (errors)
#Evaluate the model - Find the Mean Absolute Error, Mean Squared Error, Root Mean Squared Error
print('MAE: ', metrics.mean_absolute_error(y_test, predictions))
print('MSE: ', metrics.mean_squared_error(y_test, predictions))
print('RMSE : ', np.sqrt(metrics.mean_absolute_error(y_test, predictions)))
```

```python
#Explained Variance Score (R^2)
print('EVS R2: ', metrics.explained_variance_score(y_test, predictions))
```

```python
#the above variance explains 100% fit model
#Now check the residuals to ensure the good model fitness
#plot a historgam and see that it is normally distributed
sns.distplot((y_test-predictions), bins=50)
```
