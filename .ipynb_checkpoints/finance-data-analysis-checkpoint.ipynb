```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#get stock info for the BOA, Citibank, JMC, Morgan Stanely, Goldman Sachs, Wells Fargo from Jan 1, 2006 to Jan 1 2016
#Use the stock ticker name to get the bank data
startDate = datetime.datetime(2006,1,1)
endDate = datetime.datetime(2016,1,1)
BAC = data.DataReader('BAC', 'yahoo', startDate, endDate)
#Citi
C = data.DataReader('C', 'yahoo', startDate, endDate)
#JMC
JPM = data.DataReader('JPM', 'yahoo', startDate, endDate)
#Morgan Stanely
MS = data.DataReader('MS', 'yahoo', startDate, endDate)
#Goldman Sachs
GS = data.DataReader('GS', 'yahoo', startDate, endDate)
#Wells Fargo
WFC = data.DataReader('WFC', 'yahoo', startDate, endDate)
```

```python
BAC
```

```python
#create a list of tickers into alphabetical order
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
#concatenate all banks dataframe into a single bank_stocks dataframe with keys as tickers
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys=tickers)
```

```python
bank_stocks.head()
```

```python
#set the column name label
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
```

```python
bank_stocks.head()
```

```python
#max close price for each bank in the provided start and end dates
for tick in tickers:
    bank_stocks[tick].['Close'].max()
```

```python
#print using .xs (cross referencing)
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```
